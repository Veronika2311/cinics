# Word2vec

Файл cinics.txt, в той же папке, что и программа

```python
with open('cinics.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

Разделение текста на предложения

```python
import re

spr_text = re.split(r'[\n.?!]', text)

#print(spr_text)
```

Лемматинация всего с более-менее удобным форматом вывода. Внимание, это долго!

```python
%%time
from pymystem3 import Mystem
m = Mystem()

sents = []
for sent in spr_text:
    if sent:
        if not sent.startswith('%') and not sent.startswith(' %'):
            lemm  = m.lemmatize(sent)
            #print(lemm)
            sents.append(lemm)
```

На всякий случай - сохранение копии с таким трудом полученной переменной

```python
import copy
new_sents = copy.deepcopy(sents)
print(new_sents[10:20])
```

Очистка от пунктуации и прочего 'мусора'

```python
inp_for_w2v = []

for phrase in new_sents:
    clear_phrase = []
    for lemm in phrase:
        if lemm:
            if re.fullmatch('\w+', lemm):
                clear_phrase.append(lemm)
    clear_phrase.append('\n')
    inp_for_w2v.append(clear_phrase)
print(inp_for_w2v[10:20])
```

Запись всего в файл для w2v

```python
all_lemmas = ''
for phrase in inp_for_w2v:
    spr = ' '.join(phrase)
    all_lemmas = all_lemmas + spr

with open('cinics_w2v.txt', 'w', encoding='utf-8') as f:
    f.write(all_lemmas)
```

### Визуализация

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = ['ольга', 'я', 'сергей', 'марфуша', 'мамашев', 'докучаев']
X = model_cinics[words]
pca = PCA(n_components=2)
pca.fit(model_cinics[list(model_cinics.wv.vocab)])
coords = pca.transform(model_cinics[words])
```

```python
plt.scatter(coords[:, 0], coords[:, 1], color='red')
plt.title('Words')

for i, word in enumerate(words):
    plt.annotate(word, xy=(coords[i, 0], coords[i, 1]))
plt.show()
```

### Замена слов в предложениии

Поскольку слова в модели - леммы, и не всегда ближайшее слово той же части речи, что и исходное, то грамматической связности может не получиться

```python
phrase = 'Ольга умерла сегодня ночью'
result = []
lemm_phrase = m.lemmatize(phrase)
for lemm in lemm_phrase:
    if lemm in model_cinics.wv.vocab:
        lemm = model_cinics.wv.most_similar(lemm, topn=1)[0][0]
    result.append(lemm)
print(' '.join(result))
```
